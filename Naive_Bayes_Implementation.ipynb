```python
import numpy as np
import pandas as pd
```

```python
class Naive_Bayes:
    def __init__(self):
        self.result={};
    
    
    def fit(self, X_train, Y_train):
        
        self.result={}
        class_values=set(Y_train)
        for current_class in class_values:
            self.result["total_data"]=len(Y_train)
            self.result[current_class]={}
            current_class_rows=(Y_train==current_class)
            Y_train_current=Y_train[current_class_rows]
            X_train_current=X_train[current_class_rows]
            self.result[current_class]["total_count"]=len(Y_train_current)
            num_features=X_train.shape[1]
            for j in range(1,num_features+1):
                self.result[current_class][j]={}
                all_possible_values=set(X_train[:,j-1])
                for current_value in all_possible_values:
                    self.result[current_class][j][current_value]=(X_train_current[:,j-1]==current_value).sum()
        
    
    
    
    
    def predict(self,X_test):
        Y_preds=[]
        
        for x in X_test:
            x_class=self.predictSinglePoint(x)
            Y_preds.append(x_class)
        return Y_preds
   
    
    def predictSinglePoint(self,x):
        classes= self.result.keys()
        best_class=-1
        best_class_probablity=-1000
        first_run=True
        for current_class in classes:
            if(current_class=="total_data"):
                continue
            p_current_class=self.probablity(x,current_class)
            
            if(first_run or p_current_class>best_class_probablity):
                best_class_probablity=p_current_class
                best_class=current_class
            first_run=False
        return best_class
    
    
    def probablity(self,x,current_class):
        output = np.log(self.result[current_class]["total_count"])-np.log(self.result["total_data"])
        num_features=len(self.result[current_class].keys())-1;
        for j in range(1,num_features+1):
            xj=x[j-1]
            count_current_class_with_value_xj=self.result[current_class][j][xj]+1
            count_current_class=self.result[current_class]["total_count"]+len(self.result[current_class][j].keys())
            current_xj_probablity=np.log(count_current_class_with_value_xj)-np.log(count_current_class)
            output=output+current_xj_probablity
        return output
    
    
    
    
    
    
    def marked_column(column):
        second_limit=column.mean()
        first_limit=0.5*column.mean()
        third_limit=1.5*column.mean()
        
        for i in range(0,len(column)):
            if(column[i]<first_limit):
                column[i]=1
            elif(column[i]<second_limit):
                column[i]=2
            elif(column[i]<third_limit):
                column[i]=3
            elif(column[i]>=third_limit):
                column[i]=4
        return column
```

```python
from sklearn import datasets
iris= datasets.load_iris()
type(iris)
X=iris.data
Y=iris.target
X
```

```python
for j in range(0,X.shape[1]):
    X[:,j]=Naive_Bayes.marked_column(X[:,j])
```

```python
from sklearn import model_selection
X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
nb=Naive_Bayes()
nb.fit(X_train,Y_train)
```

```python
Y_predict=nb.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(Y_test,Y_predict))
print(confusion_matrix(Y_test,Y_predict))
```
